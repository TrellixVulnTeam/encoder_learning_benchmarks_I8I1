```python
# Make sure the "jacobian" method is working correctly
net = networks.perceptron.manifest.ctor(
    n_dim_in=2, n_dim_hidden=10,
    radius=1.0, initialisation="nef",
    neuron_type=networks.perceptron.ReLU)
xs = np.random.normal(0, 1, (3, net.n_dim_in))
np.set_printoptions(precision=6, suppress=True)
J = net.jacobian(xs)
J_num = net.jacobian(xs, numerical=True, eta=1e-9)
for k in J.keys():
    print(k)
    print(np.abs(J_num[k] - J[k]))
    print()
```
